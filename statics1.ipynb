```python
import numpy as np
import seaborn as sns
```

```python
arr = np.array([2,3,4,5,6,7,8,9,12,13,16,17,23,25,27,34,37,210])
```

```python
Q1 = np.percentile(arr,25)
Q2 = np.percentile(arr,75)
Q1
```

```python
Q2
```

```python
IQR = Q2 - Q1
IQR
```

```python
UF = Q2 + (1.5 * IQR)
UF
```

```python
LF = Q1 - 1.5 * IQR
LF
```

```python
l = []
for i in arr:
    if i>= LF and i<= UF:
        l.append(i)
arr2 = np.array(l)
```

```python
print(arr2)
```

```python
sns.boxplot(x=arr)
```

```python
sns.boxplot(x= arr2)
```
